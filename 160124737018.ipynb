#**Q1: Q1: Calculate Average Credit Disbursed (BL-3)**
This question is about finding the average amount of credit disbursed across all records in the dataset.
#**Using the pandas library, the .mean() function computes this average efficiently.**
import pandas as pd
Assume df is your DataFrame and'credit_disbursed' is the column that stores the numeric amount of credit given to borrowers.

```python
# Import the pandas library for data analysis
import pandas as pd

# Load the dataset
df = pd.read_csv("credit_data.csv")

# Calculate the average of 'credit_disbursed_crores' column
average_credit = df['credit_disbursed_crores'].mean()

# Display the average credit disbursed
print("Q1: Average Credit Disbursed (in crores):", round(average_credit, 2))
```

# Q2: Filter Credit Data by Sector (BL-3)

```python
# Import pandas for data handling
import pandas as pd

# Load the dataset
df = pd.read_csv("credit_data.csv")

# Choose a sector to filter data (example: Agriculture)
sector_name = "Agriculture"

# Filter the dataset for the selected sector
filtered_sector = df[df['sector'] == sector_name]

# Display the sector name and corresponding filtered data
print("Q2: Credit Data for Sector:", sector_name)
print(filtered_sector)
```

# Q3: Impute Missing Credit Records (BL-4)

```python
# Import pandas for data manipulation
import pandas as pd

# Load the dataset
df = pd.read_csv("credit_data.csv")

# Check and display missing values before imputation
print("Q3: Missing values before:", df['credit_disbursed_crores'].isnull().sum())

# Fill missing values in 'credit_disbursed_crores' with the column mean
df['credit_disbursed_crores'].fillna(df['credit_disbursed_crores'].mean(), inplace=True)

# Check and display missing values after imputation
print("Q3: Missing values after:", df['credit_disbursed_crores'].isnull().sum())
```

# Q4: Group Credit Data by State and Sector (BL-4)

```python
# Import pandas for data grouping and aggregation
import pandas as pd

# Load the dataset
df = pd.read_csv("credit_data.csv")

# Group the data by 'state' and 'sector' and calculate the total credit disbursed
grouped_data = df.groupby(['state', 'sector'])['credit_disbursed_crores'].sum().reset_index()

# Display grouped data results
print("Q4: Grouped Credit Data by State and Sector:")
print(grouped_data.head())
```

# Q5: Visualize Credit Distribution Trends (BL-5)

```python
# Import pandas and matplotlib for data visualization
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("credit_data.csv")

# Calculate the average credit disbursed for each sector
avg_credit_by_sector = df.groupby('sector')['credit_disbursed_crores'].mean().sort_values()

# Create a bar chart to visualize average credit by sector
plt.figure(figsize=(8,5))
avg_credit_by_sector.plot(kind='bar', color='skyblue', edgecolor='black')

# Add chart title and labels
plt.title("Q5: Average Credit Disbursed by Sector")
plt.xlabel("Sector")
plt.ylabel("Average Credit (in Crores)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
